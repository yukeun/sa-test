```python
import numpy as np
import librosa
import scipy
import matplotlib as mpl           # 그래프를 그리는 패키지
import matplotlib.pyplot as plt
import librosa.display
import os
from glob import glob
import IPython
%matplotlib inline
```

```python
file_list = glob('./wav/*.wav') # 디렉토리 내 wav 파일 목록 리스트화
print(file_list)
```

```python
for i in file_list:
    filename = os.path.basename(i) # 이름 추출
    data,sr = librosa.load(i, sr = None)
    time_axis = np.linspace(0, len(data)/sr, len(data))
    plt.figure(figsize = (10,4))
    plt.plot(time_axis, data)
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.title("Waveform of {}".format(filename[:-4]))
```
